```python
import pandas as pd
import numpy as np
```

```python
#load data 
data = pd.read_csv('lab 2-dataset\crx.data', header = None)
data.head()
```

```python
#create some variables names for our dataset
varnames = ['A'+str(s) for s in range(1,17)]
```

```python
#add columns names 
data.columns = varnames
```

```python
#replace ? with np,nan
data = data.replace('?', np.nan)
```

```python
#re-cast some variables types to its correct type
data['A2'] = data['A2'].astype('float')
```

```python
#encode target to binary 
data['A16'] =data['A16'].map({'+':1, '-':0})
```

```python
data.info()  #data types of columns
```

```python
data.head(15) #data.tail() #data.tail(10)
```

```python
#save the data
data.to_csv('creditApproval.csv',index=False)
```

```python
#find the categorical variables
cat_columns = [c for c in data.columns if data[c].dtypes=='O']
data[cat_columns].head(10) 
```
